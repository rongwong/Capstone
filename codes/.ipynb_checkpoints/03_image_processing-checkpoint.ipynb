```python
def image_to_df_final(path_in, preprocess_name):
    """converts images to df format in df_final"""
    
    #nested function to append file values to list for input into df
    def append_to_list(paths, emotion_type, emotion_category, preprocess_name):
        folder = os.listdir(paths)
        for file in folder:
            if file == '.DS_Store':
                pass
            else:
                file_name.append(file.split('.')[0])
                img_ext.append(file.split('.')[1])
                full_name.append(file)
                emotion.append(emotion_type)
                emotion_cat.append(emotion_category)
                preproccess.append(preprocess_name)
                image_x = plt.imread(paths+"/"+file)
                image_data.append(image_x)
                image_height.append(image_x.shape[0])
                image_width.append(image_x.shape[1])
                image_pixel.append(image_x.shape[0] * image_x.shape[1])
                try:
                    image_channel.append(image_x.shape[2])
                except:
                    image_channel.append(1)
                


    #list for input into df
    file_name = []
    img_ext = []
    full_name = []
    emotion = []
    emotion_cat = []
    preproccess = []
    image_data = []
    image_height = []
    image_width = []
    image_pixel = []
    image_channel = []
    target = []

    #loops through all folders to get data
    for folders in path_in:
#         for paths in lists:
        if folders[-3] == "g":   
            append_to_list(folders, "angry", 0, preprocess_name)
        elif folders[-3] == "p": 
            append_to_list(folders, "happy", 1, preprocess_name)
        elif folders[-3] == "r": 
            append_to_list(folders, "neutral", 2, preprocess_name)                
        elif folders[-3] == "s": 
            append_to_list(folders, "sad", 3, preprocess_name)                
                
                             
    
    global df_final
    
    #creates df_images
    df_final = pd.DataFrame(list(zip(preproccess, emotion, emotion_cat, file_name, img_ext, full_name,
                                      image_pixel, image_height, image_width, image_channel, image_data)), 
                       columns =['preproccess', 'emotion', 'emotion_cat', 'image_name', 'file_ext','file_name',
                                 'image_pixel', 'image_height', 'image_width', 'image_channels', 'image_data'])

    print("Append to list: Done")
```

```python
#combine 3 datatsets
def all_to_train(loop_list,  end_directory):
    """moves images from folders into another folder"""
    img_num = 0
    for paths in loop_list:
        for image in paths:
            if image[-3] == "p":
                copyfile(image, end_directory %(img_num, "png"))
                img_num +=1
            if image[-3] == "j":
                copyfile(image, end_directory %(img_num, "jpg"))
                img_num +=1
            
```

```python
#crops, greyscale and scale to match FER-2013

#intantiate cropper
cropper = Cropper(width=48, height=48, face_percent=100)

def process_images(path_in, path_out):
    """crops faces and converts them into greyscale"""
    img_num=10000
    count_skipped=0
    count_success = 0
    for img in path_in:
        
        cropped_array = cropper.crop(img)
        
        if cropped_array is None:
            count_skipped +=1
            print(f"Skipped: {count_skipped} - {img}")
            
            continue
            
        else:
            cropped_image = Image.fromarray(cropped_array)
        
            bw_cropped = cropped_image.convert('L')

            if img[-3] == "p":
                bw_cropped.save(path_out %(img_num, "png"))
                img_num +=1
                count_success +=1
            if img[-3] == "j":
                bw_cropped.save(path_out %(img_num, "jpg"))
                img_num +=1
                count_success +=1
                
    print('----')            
    print(f"Successful: {count_success} images")
    print(f"Skipped: {count_skipped} images")
    print(f"Loss %: {round((count_skipped/(count_success+count_skipped)),2)*100}%")
    print(" ")
    print(" ")


```

```python
#merges FER-2013 into the processed dataset

def fer_to_train(fer_paths,  end_directory):
    """moves images from fer2013 folders into destination folder"""
    img_num = 1
    for image in fer_paths:

        if image[-3] == "p":
            copyfile(image, end_directory %(img_num, "png"))
            img_num +=1
        if image[-3] == "j":
            copyfile(image, end_directory %(img_num, "jpg"))
            img_num +=1
            
    print(f"Images moved: {img_num} images")            
    
```

```python
def convert_ae(path_folder):
    """converts images in folder to Adaptive Exposure (0.3)"""
    count_skipped= 0
    count_success= 0
    for folders in path_folder:
        for path in glob.glob("%s/*" %folders):


            #read
            image_read = cv2.imread(path)

            #adaptive exposure(ae)
            try:
                adpt_eq_img = exposure.equalize_adapthist(image_read, clip_limit=0.03)
#                 adpt_eq_img = adpt_eq_img.flatten()
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass
            
            #save
            plt.imsave(path, adpt_eq_img)
            count_success += 1

    print('----')            
    print(f"Successful: {count_success} images")
    print(f"Skipped: {count_skipped} images")
    print(f"Loss %: {round((count_skipped/(count_success+count_skipped)),2)*100}%")
    print(" ")
    print(" ")

```

```python
def convert_aee(path_folder):
    """converts images in folder to Adaptive Exposure (0.3) + Equalized"""
    count_skipped= 0
    count_success= 0
    for folders in path_folder:
        for path in glob.glob("%s/*" %folders):


            #read
            image_read = cv2.imread(path)

            #adaptive exposure(ae)
            try:
                adpt_eq_img = exposure.equalize_adapthist(image_read, clip_limit=0.03)
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass
            
            #ae + equalize(e)
            try:
                eae_img = exposure.equalize_hist(adpt_eq_img)
#                 eae_img = eae_img.flatten()
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass
            
            #save
            plt.imsave(path, eae_img)
            count_success += 1

    print('----')            
    print(f"Successful: {count_success} images")
    print(f"Skipped: {count_skipped} images")
    print(f"Loss %: {round((count_skipped/(count_success+count_skipped)),2)*100}%")
    print(" ")
    print(" ")
```

```python
def convert_e(path_folder):
    """converts images in folder to Adaptive Exposure (0.3)"""
    count_skipped= 0
    count_success= 0
    for folders in path_folder:
        for path in glob.glob("%s/*" %folders):


            #read
            image_read = cv2.imread(path)

            #equalise(e)
            try:

                eq_img = exposure.equalize_hist(image_read)
#                 eq_img = eq_img.flatten()
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass
            
            #save
            plt.imsave(path, eq_img)
            count_success += 1

    print('----')            
    print(f"Successful: {count_success} images")
    print(f"Skipped: {count_skipped} images")
    print(f"Loss %: {round((count_skipped/(count_success+count_skipped)),2)*100}%")
    print(" ")
    print(" ")

```

```python
def convert_eae(path_folder):
    """converts images in folder to Adaptive Exposure (0.3)"""
    count_skipped= 0
    count_success= 0
    for folders in path_folder:
        for path in glob.glob("%s/*" %folders):
            
            #read
            image_read = cv2.imread(path)

            #equalize(e)
            try:
                eq_img = exposure.equalize_hist(image_read)  
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass
            
            #e+ae
            try:
                aee_img = exposure.equalize_adapthist(eq_img, clip_limit=0.03)
#                 aee_img - aee_img.flatten()
                
            except:
                count_skipped += 1
                print(f"Skipped: {count_skipped} - {path}")
                pass

            #save
            plt.imsave(path, aee_img)
            count_success += 1
            
    print('----')            
    print(f"Successful: {count_success} images")
    print(f"Skipped: {count_skipped} images")
    print(f"Loss %: {round((count_skipped/(count_success+count_skipped)),2)*100}%")
    print(" ")
    print(" ")
```

```python
def preview_folder(path_folder):
    plt.figure(figsize=(5,5))
    plot_no = 1
    img_no= 0 
    for folders in path_folder:
        for path in glob.glob("%s/*" %folders):
            
            if img_no < 4:

                img = plt.imread(path)
                plt.subplot(1,4,plot_no)
                plt.imshow(img, cmap="gray")
                plt.axis('off');
                plot_no += 1
                img_no += 1
            else:
                break 
```
